```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Activation
from keras.optimizers import SGD
from keras.layers import Dense
from keras.utils import np_utils

import numpy as np
import cv2
import os

def image_to_feature_vector(image,size=(32,32)):
    return cv2.resize(image,size).flatten()

data=[]
labels=[]
classes=['cat','dog']
test_cats='dogs_cats_sample_1000/dogs_cats_sample_1000/valid/cats'
test_dogs='dogs_cats_sample_1000/dogs_cats_sample_1000/valid/dogs'
train_cats='dogs_cats_sample_1000/dogs_cats_sample_1000/train/cats'
train_dogs='dogs_cats_sample_1000/dogs_cats_sample_1000/valid/cats'

test_cats_file=os.listdir('dogs_cats_sample_1000/dogs_cats_sample_1000/valid/cats')
test_dogs_file=os.listdir('dogs_cats_sample_1000/dogs_cats_sample_1000/valid/dogs')
train_cats_file=os.listdir('dogs_cats_sample_1000/dogs_cats_sample_1000/train/cats')
train_dogs_file=os.listdir('dogs_cats_sample_1000/dogs_cats_sample_1000/valid/cats')
count=0

for names in train_cats_file:
    image=cv2.imread(f'{train_cats}/{names}')
    features=image_to_feature_vector(image)
    data.append(features)
    labels.append('cat')
    count=count+1
    
    
for names in train_dogs_file:
    image=cv2.imread(f'{train_dogs}/{names}')
    features=image_to_feature_vector(image)
    data.append(features)
    labels.append('dog') 
    count=count+1





for names in test_cats_file:
    image=cv2.imread(f'{test_cats}/{names}')
    features=image_to_feature_vector(image)
    data.append(features)
    labels.append('cat')
    count=count+1
    
for names in test_dogs_file:
    image=cv2.imread(f'{test_dogs}/{names}')
    features=image_to_feature_vector(image)
    data.append(features)
    labels.append('dog')
    count=count+1
    
le=LabelEncoder()  
labels=le.fit_transform(labels)

data=np.array(data)/255.0
labels=np_utils.to_categorical(labels, 2)


(trainData, testData, trainLabels, testLabels) = train_test_split(data, labels, test_size=0.25, random_state=42)

model=Sequential()
model.add(Dense(768,input_dim=3072,activation='relu'))
model.add(Dense(384,activation='relu',kernel_initializer='uniform'))
model.add(Dense(2))
model.add(Activation('softmax'))

sgd=SGD(lr=0.1)
model.compile(loss="binary_crossentropy", optimizer=sgd,metrics=["accuracy"])
model.fit(trainData, trainLabels, epochs=50,batch_size=128,verbose=1)

(loss, accuracy) = model.evaluate(testData, testLabels,
batch_size=128, verbose=1)
print("[INFO] loss={:.4f}, accuracy: {:.4f}%".format(loss,accuracy * 100))
          
pic=cv2.imread('cat.jpg')
features=image_to_feature_vector(pic)
features=np.array([features])

probs=model.predict(features)[0]
prediction=probs.argmax(axis=0)


label = "{}: {:.2f}%".format(classes[prediction],
probs[prediction] * 100)
cv2.putText(pic, label, (10, 35), cv2.FONT_HERSHEY_SIMPLEX,1.0, (0, 255, 0), 3)
cv2.imshow("Image", pic)
            
cv2.waitKey(0)    
cv2.destroyAllWindows()
```
